```python
import pandas as pd
from pandas import ExcelWriter
```

```python
#Merging files
dfs = []
for i in range(1, 17):
    dfs.append(pd.read_excel('TC_final'+ str(i) +'.xlsx'))        
df = pd.concat(dfs)
        
df.shape
```

```python
tags = pd.DataFrame(df.tags_y)
df_final = df.drop(columns = ['tags_x', 'tags_y'])
df_final['tags'] = tags
df_final.head()
```

```python
df_final_file = df_final.drop_duplicates(subset = 'url')
print(df_final.shape)
print(df_final_file.shape)
```

```python
writer = ExcelWriter('TC_final_12357.xlsx')
df_final_file.to_excel(writer,'Sheet1')
writer.save()
```
