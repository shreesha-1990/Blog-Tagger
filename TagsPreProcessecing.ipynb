```python
import pandas as pd
from pandas import ExcelWriter
import numpy as np
import numpy as np
from collections import Counter
```

```python
df_coding = pd.read_excel('TC_final_17966.xlsx')
```

```python
df = df_coding[df_coding['tags'] != '']
df.shape
```

```python
df.columns
```

```python
df.head()
```

```python
other_cols = ['authors', 'category', 'content', 'date', 'id', 'img_src', 'section',
       'title', 'topics', 'url']
#converting all tags to lower case
df_final = pd.concat([df[other_cols], df['tags'].str.lower()], axis=1)
#splitting each tag colums
df_final = pd.concat([df_final[other_cols], df_final['tags'].str.split(',', expand=True)], axis=1)
df_final.head()
```

```python
#select only tags data frame
df_tags = df_final[[0, 1, 2, 3, 4, 5, 6, 7]]
df_tags.head()
```

```python
#Convert Data frame to nd numpy array
tags_matrix = np.asarray(df_tags)
tags_matrix.shape
```

```python
#Selecting only tags with 20 highest counts 
arr = np.asarray(df_tags)
counter = Counter(arr.flatten())
top_twenty = [tag for tag,cnt in counter.most_common(8)]
top_twenty = top_twenty[1:]
top_twenty
```

```python
#counts for top 20 tags
counter = counter.most_common(8)
dict_counter = dict(counter)
ctr = {k:v for k,v in dict_counter.items() if k != None}
ctr
```

```python
#creating 20 new columns for each tag in top twenty selected tags
#Adding 1 for column k if that tag k is present in row i 
n = 7
row_nums = []
zeros = '0' * n
tags_sparse = np.zeros((df.shape[0], n), int)
for i in range(tags_matrix.shape[0]):       
    for k, tag in enumerate(top_twenty):
        for j in range(tags_matrix.shape[1]):
            if str(tags_matrix[i][j]) in tag:
                tags_sparse[i][k] = 1
    #Checking for rows with no top 20 tags
    #Will remove these rows from Final data frame later
    all_codes = ''.join(str(z) for z in tags_sparse[i][0:n])
    if zeros in all_codes:
        row_nums.append(i)

print(tags_sparse.shape)
print(row_nums)
```

```python
df_tags.head()
```

```python
#tags data frame
tags_final_df = pd.DataFrame(tags_sparse)
tags_final_df.columns = top_twenty
tags_final_df.head()
```

```python
#Other than tags data frame
df_other = df[other_cols]
df_other.columns = other_cols
df_other.head()
```

```python
#resetting indexes
df_other.reset_index(level=0, inplace=True)
tags_final_df.reset_index(level=0, inplace=True)
print(df_other.shape)
print(tags_final_df.shape)
tags_final_df.head()
```

```python
#concatenating two data frames
df_final=pd.concat([df_other,tags_final_df],axis=1)
df_final.drop(columns = ['index','index'], inplace=True)
print(df_final.columns)
df_final.head()
```

```python
#checking for nulls in tags
pd.isnull(df_final[top_twenty]).sum()
```

```python
#Removing rows with all zeros (i.e with no top twenty tags)
df_processed = df_final.drop(df_final.index[row_nums])
print(df_final.shape)
print(df_processed.shape)
```

```python
df_processed.head()
```

```python
writer = ExcelWriter('final_data_17406.xlsx')
df_processed.to_excel(writer,'Sheet1')
writer.save()
```
